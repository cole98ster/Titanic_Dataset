## EDA - Exploratory Data Analysis

```python
df.drop('deck', axis=1 , inplace = True)
```

```python
df.dropna( inplace = True)
```

```python
df
```

```python
sns.relplot(df, x= "age", y="embarked", color = 'orange')
plt.show()
```

```python
sns.scatterplot(df['age'] )
plt.show()
```

```python
sns.displot(
            df, x= "sex",
            color = '#FFE5B4'
            )
plt.show()
```

```python
sns.boxplot(df,x='sibsp', y= 'age')
plt.savefig("Age-Siblings-Boxplot")
```

```python
sns.catplot(
            data=df,
            x= "class",
            y='survived',
            col="sex",
            kind='bar',
            )
plt.show()
```

```python
for i in df.columns:
  column = i
  sns.displot(df, x=column, color= '#FFE5B4')
  plt.show()
```

```python
df.corr()
```

```python
sns.pairplot(df, height = 2.5)
plt.show()
```

```python
for i in df.columns:
  column = i
  fig, ax = plt.subplots(figsize=(4, 4))
  sns.histplot(df, x=column, color= '#FFE5B4')
  plt.show()
```
